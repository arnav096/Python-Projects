```python
import requests
from bs4 import BeautifulSoup
import plotly
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#creating empty list for all features
brand_1=[]
price_1=[]
rating_1=[]
offrate_1=[]
memory_1=[]
display_1=[]
battery_1=[]
camera_1=[]
rew_1=[]

#passing the number of pages to the requests
num_page=input("Enter the number of pages")
for i in range(1,int(num_page)+1):
    url='https://www.flipkart.com/search?q=xiaomi+mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+str(i)
    req=requests.get(url)         #getting the url of FlipKart page
    content=BeautifulSoup(req.content,'html.parser')
   
    #reading individual content using class and div
    brand=content.find_all('div',{'class':"_4rR01T"})
    price=content.find_all('div',{'class':"_3I9_wc _27UcVY"})
    rating=content.find_all('div',{'class':"_3LWZlK"})
    offrate=content.find_all('div',class_="_3Ay6Sb")
    memory=content.find_all('div',class_="fMghEO")
    display=content.find_all('div',class_="fMghEO")
    battery=content.find_all('div',class_="fMghEO")
    camera=content.find_all('div',class_="fMghEO")
    review=content.find_all('span',class_="_2_R_DZ")
    
    #getting the results and appending it to the empty list above
    for i in brand:
        brand_1.append(i.text)
    for i in price:
        price_1.append(i.text)
    for i in rating:
        rating_1.append(i.text)
    for i in offrate:
        offrate_1.append(i.text)
    for i in review:
        rew_1.append(i.text)
    for i in memory:
        mem=i.find_all('li',class_='rgWa7D')
        m=mem[0].text     #fetching at index 0 
        memory_1.append(m)
    for i in display:
        disp=i.find_all('li',class_='rgWa7D')
        d=disp[1].text    #fetching at index 1 
        display_1.append(d)
    for i in camera:
        cam=i.find_all('li',class_='rgWa7D')
        c=cam[2].text     #fetching at index 2 
        camera_1.append(c)
    for i in battery:
        bat=i.find_all('li',class_='rgWa7D')
        b=bat[3].text     #fetching at index 3 
        battery_1.append(b)
```

```python
print("Length of each column")
print("Display: ",len(display_1))
print("Brand:   ",len(brand_1))
print("Price:   ",len(price_1))
print("Star:  ",len(rating_1))
print("OffRate: ",len(offrate_1))
print("Memory:  ",len(memory_1))
print("Camera:  ",len(camera_1))
print("Battery: ",len(battery_1))
print("Review & Ratings: ",len(rew_1))
```
